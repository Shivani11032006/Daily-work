```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Assuming your data is in a pandas DataFrame called 'df'
# If it's not, load your data into a DataFrame first

# Select the features for clustering (SAT, Top10, Accept, SFRatio, Expenses, GradRate)
X = df[['SAT', 'Top10', 'Accept', 'SFRatio', 'Expenses', 'GradRate']]

# Scale the features using StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Initialize the KMeans model with a desired number of clusters (e.g., 3)
kmeans = KMeans(n_clusters=3, random_state=42)  # You can change the number of clusters

# Fit the model to the scaled data
kmeans.fit(X_scaled)

# Get the cluster labels for each university
cluster_labels = kmeans.labels_

# Add the cluster labels to the DataFrame
df['Cluster'] = cluster_labels

# Print the DataFrame with cluster assignments
print(df)
```

```python
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with the 'Cluster' column added

# Choose two features for the scatter plot (e.g., 'SAT' and 'Accept')
x_feature = 'SAT'
y_feature = 'Accept'

# Create the scatter plot
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
plt.scatter(df[x_feature], df[y_feature], c=df['Cluster'], cmap='viridis')  # Color points by cluster
plt.xlabel(x_feature)
plt.ylabel(y_feature)
plt.title('KMeans Clustering Results')
plt.colorbar(label='Cluster')  # Add a colorbar to show cluster mapping
plt.show()
```
